```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,roc_auc_score,roc_curve,confusion_matrix, classification_report
```

```python
data = load_breast_cancer()
```

```python
x = data.data
y = data.target
```

```python
print("Classes:",data.target_names)
```

```python
x_train, x_test, y_train, y_test= train_test_split(
    x,y,test_size = 0.3
)
```

# Table1

### LogisticRegression

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(x_train, y_train)

pred_lr = lr.predict(x_test)

print("accuracy_test:" , accuracy_score(y_test,pred_lr))
pred_lr_train = lr.predict(x_train)
print("accuracy_train:" , accuracy_score(y_train,pred_lr_train))
```

###  DecisionTreeClassifier


```python
dt =  DecisionTreeClassifier()
dt.fit(x_train, y_train)

pred_dt = dt.predict(x_test)

print("accuracy_test:" , accuracy_score(y_test,pred_dt))
pred_dt_train = lr.predict(x_train)
print("accuracy_train:" , accuracy_score(y_train,pred_dt_train))
```

### KNeighborsClassifier

```python
knn =  KNeighborsClassifier()
knn.fit(x_train, y_train)

pred_knn = knn.predict(x_test)

print("accuracy_test:" , accuracy_score(y_test,pred_knn))
pred_knn_train = knn.predict(x_train)
print("accuracy_train:" , accuracy_score(y_train,pred_knn_train))
```

### GaussianNB

```python
NB =  GaussianNB()
NB.fit(x_train, y_train)

pred_NB = NB.predict(x_test)

print("accuracy_test:" , accuracy_score(y_test,pred_NB))
pred_NB_train = NB.predict(x_train)
print("accuracy_train:" , accuracy_score(y_train,pred_NB_train))
```

### MLPClassifier



```python
M = MLPClassifier()
M.fit(x_train, y_train)

pred_M = M.predict(x_test)
pred_M_train = M.predict(x_train)

print("accuracy_test:" , accuracy_score(y_test,pred_M))
print("accuracy_train:" , accuracy_score(y_train,pred_M_train))
```

### Table-C: Confusion Matrix and Error Analysis
